**Project Breakdown – Data Source**
**Simulated IoT Data for Precision Farming**

Since real IoT sensors may not be available for initial testing, we will simulate IoT sensor data to represent real-world conditions. The simulated dataset will mimic readings from soil moisture, temperature, and humidity sensors over time.

1. **Data Parameters (Sensors to Simulate)**

The system will generate data for the following key environmental factors:

*   Soil Moisture (%) – Determines water content in the soil (range: 20% to 50%).
*  Temperature (°C) – Measures environmental temperature (range: 18°C to 35°C).


*  Humidity (%) – Tracks air moisture levels (range: 40% to 80%).









2. **Data Generation Method**

Since actual sensors are not used, we will generate random sensor readings that fluctuate within realistic ranges. The data will be stored in a CSV file or database for analysis.

```python
import random
import pandas as pd
from datetime import datetime, timedelta

# Generate timestamps for 100 hours (simulated data)
timestamps = [datetime.now() - timedelta(hours=i) for i in range(100)]

# Generate random sensor readings
data = {
    "Timestamp": timestamps,
    "Soil_Moisture": [random.uniform(20, 50) for _ in range(100)],
    "Temperature": [random.uniform(18, 35) for _ in range(100)],
    "Humidity": [random.uniform(40, 80) for _ in range(100)]
}

# Create a DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv("simulated_sensor_data.csv", index=False)

print("Simulated IoT sensor data saved as 'simulated_sensor_data.csv'")
```

**Descriptive Statistics: Analyzing Data Trends for Optimal Crop Conditions**

**Objective:**

To analyze soil moisture, temperature, and humidity trends using descriptive statistics to determine optimal conditions for crop growth. This involves computing key statistical measures, identifying patterns, and detecting correlations among the environmental factors.

1. **Key Descriptive Statistics to Compute** :

**Metrics and their purposes ** :                     

1.  Mean (Average) - Shows the typical value for each parameter.
2.  Median  - Represents the central value, reducing the impact of outliers.


3.  Mode - Identifies the most frequently occurring value.	   
4. Variance -	Measures how much values deviate from the mean.



5.  Standard Deviation - Shows the spread of data points.
6.  Correlation - Finds relationships between temperature, humidity, and soil   moisture.
                                       



2. **Implementing Descriptive Statistics in Python**:

We will analyze the simulated IoT sensor data from the previous step.

```python
import pandas as pd

# Load simulated data
df = pd.read_csv("simulated_sensor_data.csv")

# Display the first few rows
print(df.head())
```

```python
# Get summary statistics for all sensor readings
stats = df.describe()
print(stats)
```

```python
# Compute correlation matrix, excluding non-numeric columns
correlation_matrix = df.drop(columns=['Timestamp']).corr()
print(correlation_matrix)
```

3. Visualizing Data Trends :

Using matplotlib and seaborn to plot trends for better insights.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Soil_Moisture'], label="Soil Moisture", color='blue')
plt.plot(df['Timestamp'], df['Temperature'], label="Temperature", color='red')
plt.plot(df['Timestamp'], df['Humidity'], label="Humidity", color='green')

plt.xlabel("Time")
plt.ylabel("Sensor Readings")
plt.title("Environmental Conditions Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Environmental Factors")
plt.show()
```

4. **Identifying Optimal Crop Conditions**

From the descriptive statistics, we derive ideal conditions for crops:

*   
Soil Moisture: Should remain between 30% and 45%.

*   Temperature: Should ideally stay between 22°C and 30°C.
*   Humidity: Optimal range 50% - 70% for crop health.

If the data shows deviations:


*   If soil moisture < 30% → Irrigation is needed.

*   If temperature > 32°C → Cooling strategies should be applied.
*   
If humidity < 50% → Dry conditions require misting.
